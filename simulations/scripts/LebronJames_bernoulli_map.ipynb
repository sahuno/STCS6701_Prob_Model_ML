```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta


#prior parameers
#how to choose prior?
# chooose basesd on expectation of prior. expectation of beta distri is the theta of making shots/ppositives
alpha_prior = 1
beta_prior = 1


lengthLeBron = len(LebronShots)
positivesLebron = np.sum(LebronShots)


# Posterior parameters
#alpha_post = alpha_prior + y
#beta_posterior = beta_prior + n - y
alpha_posterior = alpha_prior + positivesLebron 
beta_posterior = beta_prior + lengthLeBron - positivesLebron

# Generate values between 0 and 1 for p (theta, parameter of bernoulli)
p_values = np.linspace(0, 1, 1000)
#but is 1000 thetas all that you coould sample?


# Compute the PDF for each value of p
posterior_values = beta.pdf(p_values, alpha_posterior, beta_posterior) #why is  beta a pdf of thetas ?
prior_values = beta.pdf(p_values, alpha_prior, beta_prior)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(p_values, posterior_values, label=f'Beta({alpha_posterior}, {beta_posterior})', color='blue')
plt.plot(p_values, prior_values, label=f'Beta({alpha_prior}, {beta_prior})', color='red')
plt.fill_between(p_values, 0, posterior_values, color='blue', alpha=0.1)
plt.title('Posterior Distribution')
plt.xlabel('p')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

#conclusion is that whith any value of prior, the posterior is good approaximate to true distribution
```
